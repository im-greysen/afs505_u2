Using a dataset of your choice, practice creating a plot for each of these different styles:

```python
sns.set_style('darkgrid')
sns.axes_style()
sns.relplot(x = 'horsepower', y = 'mpg', data = mpg_df)
```

```python
sns.set_style('whitegrid')
sns.axes_style()
sns.relplot(x = 'horsepower', y = 'mpg', data = mpg_df)
```

```python
sns.set_style('dark')
sns.axes_style()
sns.relplot(x = 'horsepower', y = 'mpg', data = mpg_df)
```

```python
sns.set_style('white')
sns.axes_style()
sns.relplot(x = 'horsepower', y = 'mpg', data = mpg_df)
```

```python
sns.set_style('ticks')
sns.axes_style()
sns.relplot(x = 'horsepower', y = 'mpg', data = mpg_df)
```
